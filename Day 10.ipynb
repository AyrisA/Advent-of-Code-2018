Each line represents one point. Positions are given as <X, Y> pairs: X represents how far left (negative) or right (positive) the point appears, while Y represents how far up (negative) or down (positive) the point appears.

At 0 seconds, each point has the position given. Each second, each point's velocity is added to its position. So, a point with velocity <1, -2> is moving to the right, but is moving upward twice as quickly. If this point's initial position were <3, 9>, after 3 seconds, its position would become <6, 3>.

````python
from collections import defaultdict
import re
import numpy as np


def point_factory():
    return {'pos':(0,0),'vel':(0,0)}

point = defaultdict(point_factory)


i = 0
for p in pointinfo:
    pair = re.finditer('<[0-9\-\s]+,[0-9\-\s]+>',p)
    point[i]['pos'] = tuple(int(x) for x in re.sub('<*>*','',next(pair).group()).split(','))
    point[i]['vel'] = tuple(int(x) for x in re.sub('<*>*','',next(pair).group()).split(','))
    i += 1

def get_display_area(point):
    xmin = point[0]['pos'][0]
    xmax = point[0]['pos'][0]
    ymin = point[0]['pos'][0]
    ymax = point[0]['pos'][0]
    for p in point.values():
        px = p['pos'][0]
        py = p['pos'][1]
        if px < xmin:
            xmin = px
        if px > xmax:
            xmax = px
        if py < ymin:
            ymin = py
        if py > ymax:
            ymax = py
    
    width = xmax - xmin + 1
    height = ymax - ymin + 1
    return (xmin,ymin,width,height)


# Iterate steps and minimize the width/height of bounding box to find the step w/ the message!
xoff,yoff,w,h = get_display_area(point)
Wmin = w
Hmin = h
print(xoff,yoff,w,h)
c = 1
while (w <= Wmin) and (h <= Hmin):
    Wmin = w
    Hmin = h
    for p in point.values():
        x,y = p['pos']
        vx,vy = p['vel']
        p['pos'] = (x+vx,y+vy)
    
    xoff,yoff,w,h = get_display_area(point)
    c += 1
    
print(xoff,yoff,w,h)

# I'm close. Backup a couple of steps
for i in range(2):
    c -= 1
    grid[:] = '.'
    for p in point.values():
        x,y = p['pos']
        vx,vy = p['vel']
        p['pos'] = (x-vx,y-vy)
        grid[y-yoff,x-xoff] = '#'
for l in grid:
    print(''.join(l))
xoff,yoff,w,h = get_display_area(point)
print(xoff,yoff,w,h)
print("Seconds: ",c)
````
